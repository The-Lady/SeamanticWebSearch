```python
#Importing required libraries
import pandas as pd
import numpy as np
import scipy
```

Reading the CSV file


```python
data = 'final_data.csv'
df = pd.read_csv(data)
df.head()
```

```python
df=df.drop(df.columns[0], axis =1 )
```

```python
df['total_content'] = df['title'] + df['content']
```

```python
df.head()
```

Loading the pre-trained model

```python
sentences = df['total_content'].values.tolist()
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

Generating embeddings

```python
sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings
```

Checking the embeddings generated

```python
print('Sample BERT embedding vector - length', len(sentence_embeddings[0]))

print('Sample BERT embedding vector - note includes negative values', sentence_embeddings[0])
```

Testing the output of pre-trained Semantic Search Model

```python
query = 'Moderate improvement in economy' #@param {type: 'string'}

queries = [query]
query_embeddings = model.encode(queries)

number_top_matches =  8#@param {type: "number"}

print("Semantic Search Results in pre-trained models")

for query, query_embedding in zip(queries, query_embeddings):
    distances = scipy.spatial.distance.cdist([query_embedding], sentence_embeddings, "cosine")[0]


    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1])
    #print(type(results))

    print("\n-----------------\n")
    print("Query:", query)
    print("\nTop " + str(number_top_matches)+ " most similar news headlines: \n")

    for idx, distance in results[0:number_top_matches]:
        print(sentences[idx][:100], "(Cosine Score: %.4f)" % (1-distance))
```
