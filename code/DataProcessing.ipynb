Reading different datasets one by one and dropping additional columns. Renaming columns to title and content to merge all the datasets.

```python
data_1 = 'buzzfeed.csv'
df_1 = pd.read_csv(data_1)
df_1.head()
df_1.columns
```

```python
df_1 = df_1.drop(columns=['Uniq Id', 'Crawl Timestamp', 'Short Description  Sub Headline', 'Author', 'Date And Time Of Posting', 'Image Urls'], axis = 1)
df_1.head()
```

```python
df_1.columns
df_1 = df_1.rename(columns={'Title  Headline':'title','Content  Body':'content'})
df_1.head()
df_1.shape
```

```python
data_2 = 'gfg.csv'
df_2 = pd.read_csv(data_2)
df_2.head()
df_2.columns
df_2.shape
```

```python
df_2 = df_2.drop(columns= ['url', 'rating', 'tags'], axis = 1)
df_2.head()
```

```python
df_3 = pd.read_csv('medium.csv')
df_3.head()
df_3.columns
```

```python
df_3 = df_3.drop(columns =['author', 'claps', 'reading_time', 'link'], axis = 1)
df_3.head()
df_3.columns
```

```python
df_3 = df_3.rename(columns={'text':'content'})
df_3.head()
```

```python
df_4 = pd.read_csv('covidArticles.csv')
df_4.head()
df_4.columns
```

```python
df_4 = df_4.drop(columns= ['Unnamed: 0', 'Unnamed: 0.1','doi', 'date', 'xml_url', 'url'], axis = 1)
df_4.head()
```

```python
df_4 = df_4.rename(columns={'abstract':'content'})
df_4.head()
```

```python
df_5 = pd.read_csv('news.csv',sep='\t')
df_5.head()
df_5.columns
```

```python
df_5 = df_5.drop(columns = ['category', 'filename'], axis = 1)
df_5.head()
```

Removing duplicates from all the datasets

```python
df_1 = df_1.drop_duplicates()
df_2 = df_2.drop_duplicates()
df_3 = df_3.drop_duplicates()
df_4 = df_4.drop_duplicates()
df_5 = df_5.drop_duplicates()
```

Removing NA values

```python
df_1 = df_1.loc[:,:].dropna(axis=1)
df_2 = df_2.loc[:,:].dropna(axis=1)
df_3 = df_3.loc[:,:].dropna(axis=1)
df_4 = df_4.loc[:,:].dropna(axis=1)
df_5 = df_5.loc[:,:].dropna(axis=1)
```

Concatinating equal number of rows from all the datasets.

```python
final_data = pd.concat([df_1[:300], df_2[:300], df_3[:300], df_4[:300], df_5[:300]])
```
